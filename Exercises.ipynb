#### Exercise 1 - Class Inheritance

Create an Employee class that sets an employee's first name, last name, job title, salary, and email. The Employee class should have a class attribute for the raise amount set to 5% (1.05). Create a method that will apply the raise to an employee's salary.

```python
class Employee:        

    raise_amount = 1.05
    
    def __init__(self, first_name, last_name, job_title, salary):
    
        self.first_name = first_name
        self.last_name = last_name
        self.job_title = job_title
        self.salary = salary
        self.email = first_name + '.' + last_name + '@company.com'

    
    def give_raise(self):
        self.salary = int(self.salary) * self.raise_amount 
    
example_guy = Employee('Ash', 'Ketchum', 'Pokemon Trainer', 500000)
```

```python
print(example_guy.salary)
example_guy.give_raise()
print(example_guy.salary)
```

Create two more classes that inherit from the Employee class. One for Sales and one for Development. Both of these classes will have the same attributes as the Employee.<br>
- For the Sales employees, add a phone number attribute on instantiation using the super method. 
- Create a method on the Sales class that will Send a Follow Up Email. It should take in a customer name and "send" aka print a formatted email "Dear *customer*, Thank you for your interest in our product. Please let me know if you have any questions. My email is *email* or my phone number is *phone number*. Thanks, *full name*"
- Create a method on the Development class called code that will print out "*full name* is writing code".

```python
class Sales(Employee):
    def __init__(self, first_name, last_name, job_title, salary, phone_number):
        super().__init__(first_name, last_name, job_title, salary)
        self.phone_number = phone_number
        
    def follow_up(self, customer_name):
        self.customer_name = customer_name
        print(f"Dear {self.customer_name}, Thank you for your interest in our product. Please let me know if you have any questions. My email is {self.email} or my phone is {self.phone_number}. Thanks, {self.first_name} {self.last_name}")

class Development(Employee):
    def __init__(self, first_name, last_name, job_title, salary):
        super().__init__(first_name, last_name, job_title, salary)
    
    def code(self):
        print(f'{self.first_name} {self.last_name} is writing code.')
```

- Create an instance of a Sales Employee with a salary of $50,000.
- Send follow up emails to "Mike O'Neil" and "Hannah Stern"
- Give the employee a raise and print the salary

```python
salesguy1 = Sales("Joe", "Dunlap", "Salesguy",'50000', '8171115555')
salesguy1.follow_up("Mike O'Neil")
salesguy1.follow_up("Hannah Stern")

salesguy1.give_raise()
```

- Create an instance of a Development Employee with a salary of $100,000
- Write some code with this employee
- Give the employee a raise and print the salary

```python
devguy1 = Development('Mike', 'Wazowski', 'Developer Guy', 100000)

print(devguy1.__dict__)

devguy1.code()
devguy1.give_raise()
print(devguy1.salary)
```
